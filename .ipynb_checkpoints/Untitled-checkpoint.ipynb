```python
import cv2
from ultralytics import YOLO

model = YOLO("runs/detect/train/weights/best.pt") 

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(frame, conf=0.5, verbose=False)
    annotated_frame = results[0].plot()

    cv2.imshow("Gesture Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

model = YOLO("runs/detect/train/weights/best.pt")

cap = cv2.VideoCapture(0)
frame_count = 0
max_frames = 10 

while frame_count < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(frame, conf=0.25, verbose=False)
    annotated_frame = results[0].plot()

    
    plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f"Frame {frame_count+1}")
    plt.show()

    frame_count += 1

cap.release()
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

model = YOLO("runs/detect/train/weights/best.pt")

cap = cv2.VideoCapture(0)

plt.ion()  # Interactive mode باش الصور تبان بسرعة

fig, ax = plt.subplots(figsize=(10, 6))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(frame, conf=0.5, verbose=False)
    boxes = results[0].boxes
    annotated_frame = results[0].plot()

    if boxes is not None and len(boxes) > 0:
        label = results[0].names[int(boxes.cls[0])]
        print(f"Gesture Detected: {label}")

        if label == "Hello":
            print("User says Hello! 👋")
        elif label == "Yes":
            print("User says Yes! ✅")
        elif label == "No":
            print("User says No! ❌")
        elif label == "Thanks":
            print("User says Thanks! 🙏")
        elif label == "I Love You":
            print("User says I Love You! ❤️")

    # نعرض الصورة عبر matplotlib بدل imshow
    ax.clear()
    ax.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
    ax.axis('off')
    plt.pause(0.001)

    # نوقف البرنامج بضغط على الزر 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
plt.close()
```
